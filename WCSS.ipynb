```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('student_clustering.csv')
print(df)
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['cgpa'], df['iq'])
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('Scatter Plot of CGPA vs IQ')
plt.show()
```

```python
wcss=[]
for i in range(1,11):
    model=KMeans(n_clusters=i)
    model.fit_predict(df)
    wcss.append(model.inertia_)

wcss
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```

```python
x=df.to_numpy()
```

```python
X=df.iloc[:,:].values
X
```

```python
model=KMeans(n_clusters=4)
Y=model.fit_predict(X)
Y
```

```python
df["classes"]=Y
df
```

```python
plt.scatter(X[Y==0,0],X[Y==0,1],color='black',label="category 0")
plt.scatter(X[Y==1,0],X[Y==1,1],color='purple',label="category 1")
plt.scatter(X[Y==2,0],X[Y==2,1],color='orange',label="category 2")
plt.scatter(X[Y==3,0],X[Y==3,1],color='pink',label="category 3")
plt.title('K Means of student data')
plt.xlabel("CGPA")
plt.ylabel("IQ")
plt.legend()
plt.show()
```
